```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_table("data.txt", sep='\t',index_col = None, names = ['p', 'ISIZE', 'JSIZE', 'time'])
df = df.groupby(['p', 'ISIZE', 'JSIZE']).mean().reset_index()
df['S'] = df['time'].apply(lambda x : df['time'].values[0] / x)
df['E'] = df['S'] / df['p']
df
```

```python
plt.figure(figsize=(15, 8))
plt.plot(df['p'], df['S'], label='Speedup')
plt.plot(df['p'], df['E'], label='efficency')
plt.legend()
plt.savefig('experiment_plot.png')
plt.show()
```
